# **Exploratory Data Analysis**

### 1️. AGE DISTRIBUTION

```python
sns.histplot(
    data=df,
    x="User_Age",
    bins=20,
    kde=True,
    color="red"
)
plt.title("User Age Distribution")
plt.xlabel("Age")
plt.ylabel("Number of Users")
plt.show()
```

### 2. Age Group Breakdown

```python
ax = sns.countplot(
    data=df,
    x="Age_Group",
    order=df["Age_Group"].value_counts().index,
    palette="Reds"
)

for container in ax.containers:
    ax.bar_label(container)

plt.title("User Distribution by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Users")
plt.show()

```

### 3. Revenue by Subscription Type

```python
ax = sns.barplot(
    data=df,
    x="Subscription_Type",
    y="Revenue_INR",
    estimator=sum,
    palette="Reds"
)

for container in ax.containers:
    ax.bar_label(container, fmt="₹%.0f", padding=3)

plt.title("Total Revenue by Subscription Type")
plt.xlabel("Subscription Type")
plt.ylabel("Total Revenue (INR)")
plt.show()

```

### 4. Revenue Distribution

```python
avg_revenue = df["Revenue_INR"].mean()

plt.figure(figsize=(6,4))
plt.bar(["Average Revenue"], [avg_revenue], color="#DD8452")

plt.title("Average Revenue per User")
plt.ylabel("Revenue (INR)")
plt.show()
```

# **Exploratory Data Analysis (EDA) – Key Insights**

### 📊 Exploratory Data Analysis (EDA) – Key Insights

1. **User Demographics**
   - Majority of users fall in the **18–34 age range**, indicating a younger user base.
   - Engagement (watch hours) is highest among **25–34 age group**, making them the most valuable segment.

2. **Engagement Patterns**
   - Users with higher **watch hours** and **completion percentage** show significantly lower churn.
   - Power users (top engagement segment) contribute disproportionately to total watch time.

3. **Churn Analysis**
   - Churned users have noticeably **lower customer satisfaction scores** compared to retained users.
   - Lower engagement and dissatisfaction are strong early indicators of churn risk.

4. **Revenue Insights**
   - Premium subscription plans contribute the **highest share of total revenue**.
   - Revenue is skewed, with a small group of users contributing a large portion of total revenue.

5. **Content Performance**
   - Certain genres show **higher completion rates**, indicating stronger content engagement.
   - High-rated shows with low completion suggest content mismatch or pacing issues.

6. **Device Usage**
   - Most users consume content on **mobile and smart TV devices**.
   - Viewing device choice impacts engagement and churn behavior.

7. **Overall Business Implications**
   - Improving content quality and engagement for low-satisfaction users could significantly reduce churn.
   - Targeted retention strategies for high-risk segments can protect revenue growth.
